##### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data_set1 = pd.read_csv('uber-raw-data-apr14.csv')
data_set2 = pd.read_csv('uber-raw-data-may14.csv')
data_set3 = pd.read_csv('uber-raw-data-jun14.csv')
data_set4 = pd.read_csv('uber-raw-data-jul14.csv')
data_set5 = pd.read_csv('uber-raw-data-aug14.csv')
data_set6 = pd.read_csv('uber-raw-data-sep14.csv')
```

##### convert datetime sting type to datetime format

```python
from datetime import datetime
```

```python
data_set['Date/Time'] = pd.to_datetime(data_set['Date/Time'])
type(data_set['Date/Time'])
```

```python
data_set['month'] = data_set['Date/Time'].dt.month
data_set['day'] = data_set['Date/Time'].dt.day
data_set['hour'] = data_set['Date/Time'].dt.hour
data_set['minute'] = data_set['Date/Time'].dt.minute
data_set['week'] = data_set['Date/Time'].dt.week
data_set['weekday_name'] = data_set['Date/Time'].dt.weekday
```
